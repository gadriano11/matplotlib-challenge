### Analysis

- Analysing Pymaceutical dataset, we explored various aspects of the data, conducted statistical analyses, and generated visualisations to gain insights into the effects of different drug regimens on tumor volumes in mice. 

- The correlation coefficient between mouse weight and average tumor volume for the Capomulin regimen is 0.84. This indicates a strong positive correlation between mouse weight and tumor volume. As the mouse weight increases, the average tumor volume tends to increase as well. 

- The coefficient of determination (R-squared) is 0.71. This value suggests that approximately 71% of the variability in average tumor volume can be explained by changes in mouse weight. It indicates a moderately strong linear relationship between these two variables.

- There exists a potential outlier data in Infubinol drug regimen.

- In the majority of the test sample for Capomulin drug regimen, there is a trending decrease in Tumor Volume vs. Time Point for Mouse on the Capomulin drug regimen. There are some samples within the group that did not show this trend and may need further analysis to determine the cause.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import random

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata, study_results, on='Mouse ID', how='right')
combined_data = combined_data[['Mouse ID', 
                               'Timepoint', 
                               'Tumor Volume (mm3)', 
                               'Metastatic Sites', 
                               'Drug Regimen', 
                               'Sex', 
                               'Age_months', 
                               'Weight (g)']]
combined_data
```

```python
# Checking the number of mice.
number_mice = combined_data['Mouse ID'].nunique()
print(f'There are {number_mice} mice in the study')
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
combined_data.set_index(['Mouse ID', 'Timepoint'])
combined_data.head()

# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_mice_ID = combined_data.loc[combined_data.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
print(f'The duplicated mice ID is: {(duplicated_mice_ID[0])}')

# Check how many times the duplicate ID appeared
duplicate_mouse_id = duplicated_mice_ID[0]
count = combined_data[combined_data['Mouse ID'] == duplicate_mouse_id]['Mouse ID'].count()
print(f'The duplicated mice ID appeared: {count} times')

# Check how many will have to be removed
count_duplicates = combined_data.duplicated(subset=['Mouse ID', 'Timepoint']).sum()
print(f'There are {count_duplicates} occurence of {(duplicated_mice_ID[0])} that needs to be removed basing on Mouse ID and Timepoint identifier.')
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicated_mice_info =  combined_data[combined_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
duplicated_mice_info
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# Five duplicates will be removed.

cleaned_combined_data = combined_data.drop_duplicates(subset=['Mouse ID', 'Timepoint'])
cleaned_combined_data
```

```python
# Checking the number of mice in the clean DataFrame.
# It would still show the same number of mice as we only removed the duplicated ones

number_mice_clean_data = cleaned_combined_data['Mouse ID'].nunique()
print(f'There are still {number_mice_clean_data} mice in the study')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

grouped_data = cleaned_combined_data.groupby("Drug Regimen")

# Calculate mean, median, variance, standard deviation, and SEM  of the tumor volume for each regimen
mean_tumor_volume = grouped_data["Tumor Volume (mm3)"].mean()
median_tumor_volume = grouped_data["Tumor Volume (mm3)"].median()
variance_tumor_volume = grouped_data["Tumor Volume (mm3)"].var()
std_deviation_tumor_volume = grouped_data["Tumor Volume (mm3)"].std()
sem_tumor_volume = grouped_data["Tumor Volume (mm3)"].sem()

# Create a summary DataFrame
summary_statistics = pd.DataFrame({
    "Mean Tumor Volume": mean_tumor_volume,
    "Median Tumor Volume": median_tumor_volume,
    "Tumor Volume Variance": variance_tumor_volume,
    "Tumor Volume Std. Dev.": std_deviation_tumor_volume,
    "Tumor Volume SEM": sem_tumor_volume
})

summary_statistics
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_statistics_alt = cleaned_combined_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"]).rename(columns={"mean": "Mean Tumor Volume", "median": "Median Tumor Volume", "var": "Tumor Volume Variance", "std": "Tumor Volume Std. Dev.", "sem": "Tumor Volume SEM"})
summary_statistics_alt
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
max_timepoints = cleaned_combined_data.groupby('Mouse ID')['Timepoint'].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_data = pd.merge(cleaned_combined_data, max_timepoints, on=['Mouse ID', 'Timepoint'], how='right')
final_tumor_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
selected_regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
final_tumor_volumes = []
outliers = []

# Loop through each selected regimen and calculate the final tumor volumes
for regimen in selected_regimens:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    regimen_data = final_tumor_data[final_tumor_data['Drug Regimen'] == regimen]['Tumor Volume (mm3)']
    # add subset 
    final_tumor_volumes.append(regimen_data)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = regimen_data.quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    median = quartiles[0.5]
    iqr = upper_q - lower_q
    lower_bound = lower_q - (1.5 * iqr)
    upper_bound = upper_q + (1.5 * iqr)
    min_value = regimen_data.min()
    max_value = regimen_data.max()
    # Determine outliers using upper and lower bounds
    regimen_outliers = regimen_data[(regimen_data < lower_bound) | (regimen_data > upper_bound)]
    # Store information in a dictionary
    outliers.append({
        'Drug': regimen,
        'Median': median,
        'Min': min_value,
        'Max': max_value,
        'Lower Quartile': lower_q,
        'Upper Quartile': upper_q,
        'IQR': iqr,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Potential Outliers': ', '.join(map(str, regimen_outliers))
    })

# Create a new dataframe
outliers_df = pd.DataFrame(outliers)
outliers_df.set_index("Drug")
outliers_df
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(final_tumor_volumes, labels=selected_regimens, flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 8})
plt.title('Final Tumor Volume Distribution by Treatment Regimen')
plt.xlabel('Treatment Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Filter data for the Capomulin regimen mice
capomulin_data = cleaned_combined_data[cleaned_combined_data['Drug Regimen'] == 'Capomulin']

# Randomly select a mouse ID from the Capomulin regimen
random_mouse_cap_id = random.choice(capomulin_data['Mouse ID'].unique())
random_mouse_cap_data = capomulin_data[capomulin_data['Mouse ID'] == random_mouse_cap_id]
print(f'The random mouse selected is ID: {random_mouse_cap_id}')

# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
plt.plot(random_mouse_cap_data['Timepoint'], random_mouse_cap_data['Tumor Volume (mm3)'], marker='o', color='blue', linestyle='-')
plt.title(f'Tumor Volume vs. Time Point for Mouse {random_mouse_cap_id} on Capomulin Regimen')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# Calculate the average observed tumor volume for each mouse on Capomulin
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
mouse_weights = capomulin_data.groupby('Mouse ID')['Weight (g)'].first()

# Create a scatter plot
plt.scatter(mouse_weights, average_tumor_volume, marker='o', facecolors='blue', edgecolors='black', alpha = 0.5)
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(True)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(mouse_weights, average_tumor_volume)
print(f"Correlation Coefficient: {round(correlation[0],2)}")

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(mouse_weights, average_tumor_volume)
regress_values = mouse_weights * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Create a scatter plot with regression line
plt.scatter(mouse_weights, average_tumor_volume, marker='o', facecolors='blue', edgecolors='black', alpha = 0.5)
plt.plot(mouse_weights,regress_values,color ='red')
plt.annotate(line_eq,(22,36),fontsize=10,color="red")
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
print(f"The r-squared is: {r_value**2}")
plt.grid(True)
plt.show()

```
